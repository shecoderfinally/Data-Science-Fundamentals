```python
import csv
import numpy as np
import pandas as pd
from collections import Counter
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn')
```

```python
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    row = next(csv_reader)
    # Output is ordered dictionary.Keys are headers in csv file and values are responses for that particular person.
    print(row)
    print()
    print(row['LanguagesWorkedWith'].split(';'))
```

```python
# This will split all the languages and print them from 1st row
print(row['LanguagesWorkedWith'].split(';'))
print()       
```

```python
# From the previous step, the current value of row is the first line of dataset
print(row)
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    language_counter=Counter()

    for row in csv_reader:
        # we applied this to every row in csv_reader with help of for loop, so it does this for every row.
        #It will return a comma separated list of languages name for every row that get updated into the langauge_counter variable.
        # So after every iteration, the ;anguage names and their counts get updated.
        language_counter.update(row['LanguagesWorkedWith'].split(';'))

# After running on entire data, the final set of languages with their counts:
# In counter the keys with most responses are listed first, followed by second highest responses   
print(language_counter) 
print()       

# To print only the most common (say 15 most common) we write
print(language_counter.most_common(15))
print()

languages = []

popularity= []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
    
print(languages)
print()
print(popularity)
```

```python
# Vertical Bar Chart
plt.bar(languages,popularity,width=0.1)
plt.title("Most Popular Languages")
plt.tight_layout()
plt.xlabel('Programming Languages')
plt.ylabel('The number of people who use')
```

### Remember to switch your x & y axis labels when switching from plt.bar and plt.barh

```python
# Horizontal Bar Chart
plt.barh(languages,popularity)

# will give error because of width attribute
#plt.barh(languages,popularity, width=0.5)

plt.title("Most Popular Languages")
plt.tight_layout()
plt.ylabel('Programming Languages')
plt.xlabel('The number of people who use')
```

You can give width attribute while plotting plt.bar but not while plotting plt.barh
In the above barh graph the most popular language is at the bottom, to reverse that we can just just reverse the list we pass to the plot function.

```python
# Horizontal Bar Chart
languages.reverse()
popularity.reverse()
plt.barh(languages,popularity)

# will give error because of width attribute
#plt.barh(languages,popularity, width=0.5)

plt.title("Most Popular Languages")
plt.tight_layout()
plt.ylabel('Programming Languages')
plt.xlabel('The number of people who use')
```

```python
# Reading data using Pandas read_csv method
df=pd.read_csv('data.csv')
df.reset_index(drop=True,inplace=True)
print(df.head())
print()

ids = df['Responder_id']
lang_responses =df['LanguagesWorkedWith']

language_counter=Counter()

for response in lang_responses:
    language_counter.update(response.split(';'))
    
languages=[]
popularity=[]

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

print(languages)
print()
print(popularity)
```
