# Pie Charts

#### They show the proportion each type contributes to the whole.Our chart can be thought of as a pie with each category  being a slice of the pie.The larger the size, the greater its contribution.Our values need to add up to 100% ,the plot adjusts accordingly.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
# Slices add up to 100
plt.title('My Awesome Pie Chart')
slices = [60,40]
labels= ['Sixty','Forty']
plt.pie(slices, labels=labels)

plt.tight_layout()
plt.show()
```

```python
#Slicesdo not add up to 100
plt.title('My Awesome Pie Chart')
slices = [120,80]
labels= ['Sixty','Forty']
plt.pie(slices, labels=labels)

plt.tight_layout()
plt.show()
```

```python
# To add separators between 2 slices
plt.title('My Awesome Pie Chart')
slices = [120,80,30,20]
labels= ['Sixty','Forty','Extra1','Extra2']\

# Specifying custom colors using hdefault color names
colors=['blue','red','yellow','green']
plt.pie(slices, labels=labels,colors= colors, wedgeprops={'edgecolor': 'Black'})

plt.tight_layout()
plt.show()
```

```python
# Using Proper Data

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']

plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'})
plt.tight_layout
plt.show()
```

If you have a lot of data then it is better to use barchart instead of piechart.So in this case we are going to use only top 5 programming languages.

```python
# Using Proper Data

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'})
plt.tight_layout
plt.show()
```

```python
# To lay emphasis on a particular slice

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

# 0 would keep the value as it is, any value other than 0 will repreent the fraction of the radius that we woudl like to explode the value
# Values you can try for explode:
# explode= [0]
# explode = [0,0,0,0.5,0]
explode = [0,0,0,0.1,0]

plt.pie(slices,labels=labels, explode =explode ,wedgeprops={'edgecolor':'black'})
plt.tight_layout
plt.show()
```

```python
# To give our chart a shadow, we pass and argument shadow and set it equal to true

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

# 0 would keep the value as it is, any value other than 0 will repreent the fraction of the radius that we woudl like to explode the value
# Values you can try for explode:
# explode= [0]
# explode = [0,0,0,0.5,0]
explode = [0,0,0,0.1,0]

plt.pie(slices,labels=labels, explode =explode, shadow='True',wedgeprops={'edgecolor':'black'})
plt.tight_layout
plt.show()
```

```python
# To give starting angle, say 90

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

# 0 would keep the value as it is, any value other than 0 will repreent the fraction of the radius that we woudl like to explode the value
# Values you can try for explode:
# explode= [0]
# explode = [0,0,0,0.5,0]
explode = [0,0,0,0.1,0]

plt.pie(slices,labels=labels, explode =explode, shadow='True', startangle=90, wedgeprops={'edgecolor':'black'})
plt.tight_layout
plt.show()
```

```python
# To display the percentage each slice is taking up,

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

# 0 would keep the value as it is, any value other than 0 will repreent the fraction of the radius that we woudl like to explode the value
# Values you can try for explode:
# explode= [0]
# explode = [0,0,0,0.5,0]
explode = [0,0,0,0.1,0]

plt.pie(slices,labels=labels, explode =explode, shadow='True', startangle=90, autopct='%1.1f%%',wedgeprops={'edgecolor':'black'})
plt.tight_layout
plt.show()
```

```python
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']
sli = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]

labels.reverse()
sli.reverse()

plt.barh(labels,sli)
plt.title('Language Statistics')
plt.xlabel('Frequency')
plt.ylabel('Languages')
plt.tight_layout
plt.show()
```
